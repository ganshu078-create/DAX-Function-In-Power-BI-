#DAX Function In Power BI Assignment Questions

Dataset Description

The dataset consists of the following columns:

| Column Name     | Description               |
| --------------- | ------------------------- |
| OrderID         | Unique Order Identifier   |
| OrderDate       | Date of Purchase          |
| ProductID       | Unique Product Identifier |
| ProductCategory | Product Category          |
| Quantity        | Number of Units Sold      |
| UnitPrice       | Price Per Unit            |
| Discount        | Discount Percentage       |
| SalesAmount     | Total Billed Amount       |
| Region          | Sales Region              |
| SalesChannel    | Online / Mobile App       |


**Question 1. Create a calculated column TotalPrice using Quantity and Unit Price.**

- DAX Formula

TotalPrice =
AmazonSales[Quantity] * AmazonSales[UnitPrice]

Explanation

This calculated column multiplies the quantity sold by the unit price to determine the total price before applying any discount.

Output

TotalPrice = Quantity × UnitPrice

**Question 2. Create a calculated column DiscountedPrice.**

- DAX Formula

DiscountedPrice =
AmazonSales[TotalPrice] * (1 - AmazonSales[Discount])

Explanation

This calculated column calculates the final amount after applying the discount percentage on the total price.

Output

DiscountedPrice = TotalPrice × (1 - Discount)

**Question 3. Create a measure Average Order Value (AOV).**

- DAX Formula

Average Order Value (AOV) =
AVERAGE(AmazonSales[SalesAmount])

Explanation

Average Order Value (AOV) shows the average revenue generated per order.

Result

Average Order Value (AOV) = 2516.08

**Question 4: Create a Measure for Year-To-Date (YTD) Sales**

- DAX Formula

YTD Sales =

TOTALYTD
(

    SUM(AmazonSales[SalesAmount]),

    AmazonSales[OrderDate]
)

Explanation

This measure calculates cumulative sales from the beginning of the year up to the selected date.

Business Use

Management can use this KPI to track yearly sales performance and growth trends.

**Question 5. Using DAX, create a measure to rank each product according to its Total Sales, with the highest-selling
product ranked first**

- DAX Formula
Product Rank =

RANKX(

    ALL(AmazonSales[ProductID]),
    CALCULATE(SUM(AmazonSales[SalesAmount])),
    ,
    DESC,
    DENSE
)

Explanation

This measure ranks products based on their total sales.

Highest sales product receives Rank 1.
Ranking is performed in descending order.
Dense ranking avoids gaps in rank numbers.
Business Use

This helps identify top-performing and low-performing products.

**Question 6. Create a calculated column classifying orders as “High Value” if sales > 2000, else “Low Value”.**

- DAX Formula

Order Category =
IF(

    AmazonSales[SalesAmount] > 2000,
    "High Value",
    "Low Value"
)

Explanation

Orders are categorized based on their sales amount:

Sales Amount greater than 2000 → High Value
Sales Amount less than or equal to 2000 → Low Value

Business Use

This classification helps management focus on high-revenue customers and orders.

**Question 7. Create a measure to calculate total sales from the Online channel only.**

- DAX Formula

Online Sales =
CALCULATE(

    SUM(AmazonSales[SalesAmount]),
    AmazonSales[SalesChannel] = "Online"
)

Explanation

This measure calculates sales generated only through the Online channel.

Result

Online Sales = 677,525.24

**Question 8. Create a measure called "Date Difference" that calculates the number of days between the latest  Order Date and today's date using the DATEDIFF function**

- DAX Formula
Date Difference =

DATEDIFF(

    MAX(AmazonSales[OrderDate]),
    TODAY(),
    DAY
)

Explanation

This measure calculates the number of days between the latest order date in the dataset and today's date.

Result

Latest Order Date = 15-May-2023

Date Difference = 1112 Days
